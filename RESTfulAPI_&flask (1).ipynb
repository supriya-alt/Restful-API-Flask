# Question 1 : What is a RESTful API?

  -> RESTful API(Representation State Transfer API) is basically a set of rules for desigining web services. It uses standard HTTP methods like GET,POST,PUT,DELETE to allow clients(like apps  or browsers) to interact with resources,which are typically represented as URLs. A RESTful API is stateless ans usually returns data in a format like jJSON or XML.this makes it easy to build scalable,flexible,and interoperable web services.

# Question 10: How do you structure a Flask app using Blueprints?

  --> In flask using blueprints allows you to organize application into reusable components. First, create a separate python file for each logical part of your app. in each file you define a blueprint with Blueprint(), then and routes to that blueprint. Afterwards, in our  main application file, we import and register each  blueprint using app.register_blueprint().
